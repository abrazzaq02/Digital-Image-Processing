```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('dogii.webp', 0)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python

hist = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

```python

plt.hist(img.ravel(), bins=256, range=(0, 256))
plt.title("Grayscale Histogram")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")

plt.xlim([0, 256])
plt.show()
```

```python

color_img = cv2.imread('dogii.webp')

if color_img is None:
    print("Error: Could not load the color image. Please check the path.")
else:
    # Split the image into color channels
    b, g, r = cv2.split(color_img)

    # Display the color image
    plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
    plt.title("Color Image")
    plt.axis('off')
    plt.show()
```

```python

hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

plt.figure(figsize=(10, 6))
plt.title("Color Image Histogram")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")
plt.plot(hist_b, color='blue', label='Blue channel')
plt.plot(hist_g, color='green', label='Green channel')
plt.plot(hist_r, color='red', label='Red channel')
plt.xlim([0, 256])
plt.legend()
plt.show()
```

```python

img_equalized = cv2.equalizeHist(img)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_equalized, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

hist_equalized = cv2.calcHist([img_equalized], [0], None, [256], [0, 256])
plt.subplot(1, 2, 2)
plt.plot(hist_equalized)
plt.title("Equalized Histogram")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python


img = cv2.imread('dogii.webp', cv2.IMREAD_GRAYSCALE)
eq_img = cv2.equalizeHist(img)


hist1 = cv2.calcHist([img],[0],None,[32],[0,128])
hist2 = cv2.calcHist([eq_img],[0],None,[32],[0,128])
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.plot(hist1); plt.title('Original (32 bins, 0128)')
plt.subplot(1,2,2); plt.plot(hist2); plt.title('Equalized')
plt.show()

hist1 = cv2.calcHist([img],[0],None,[128],[50,200])
hist2 = cv2.calcHist([eq_img],[0],None,[128],[50,200])
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.plot(hist1); plt.title('Original (128 bins, 50200)')
plt.subplot(1,2,2); plt.plot(hist2); plt.title('Equalized')
plt.show()


hist1 = cv2.calcHist([img],[0],None,[512],[100,256])
hist2 = cv2.calcHist([eq_img],[0],None,[512],[100,256])
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.plot(hist1); plt.title('Original (512 bins, 100256)')
plt.subplot(1,2,2); plt.plot(hist2); plt.title('Equalized')
plt.show()
```
