```python

color_img = cv2.imread('Miao.jpg')

if color_img is None:
    print("Error: Could not load the color image. Please check the path.")
else:
    # Split the image into color channels
    b, g, r = cv2.split(color_img)

    # Display the color image
    plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
    plt.title("Color Image")
    plt.axis('off')
    plt.show()
```

```python
# Compute the histograms for each color channel
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

# Plot the histograms
plt.figure(figsize=(10, 6))
plt.title("Color Image Histogram")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")
plt.plot(hist_b, color='blue', label='Blue channel')
plt.plot(hist_g, color='green', label='Green channel')
plt.plot(hist_r, color='red', label='Red channel')
plt.xlim([0, 256])
plt.legend()
plt.show()
```

```python
# Apply histogram equalization
img_equalized = cv2.equalizeHist(img)

# Display the equalized image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_equalized, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

# Compute and display the histogram of the equalized image
hist_equalized = cv2.calcHist([img_equalized], [0], None, [256], [0, 256])
plt.subplot(1, 2, 2)
plt.plot(hist_equalized)
plt.title("Equalized Histogram")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Load the image ---
# Try different images: dark, bright, or low contrast
img = cv2.imread('your_image.jpg', cv2.IMREAD_GRAYSCALE)

# --- 2. Function to compute histogram ---
def compute_histogram(image, bins=256, range_vals=(0, 256)):
    hist = cv2.calcHist([image], [0], None, [bins], range_vals)
    hist = hist.flatten()
    return hist

# --- 3. Plot function ---
def plot_histograms(original, equalized, bins, range_vals):
    plt.figure(figsize=(12, 6))

    # Plot Original Image and Histogram
    plt.subplot(2, 2, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')

    plt.subplot(2, 2, 2)
    plt.plot(compute_histogram(original, bins, range_vals), color='gray')
    plt.title(f'Original Histogram (bins={bins}, range={range_vals})')

    # Plot Equalized Image and Histogram
    plt.subplot(2, 2, 3)
    plt.imshow(equalized, cmap='gray')
    plt.title('Equalized Image')

    plt.subplot(2, 2, 4)
    plt.plot(compute_histogram(equalized, bins, range_vals), color='black')
    plt.title('Equalized Histogram')

    plt.tight_layout()
    plt.show()

# --- 4. Apply histogram equalization ---
equalized_img = cv2.equalizeHist(img)

# --- 5. Experimentation Section ---
# Try different settings:
bins_list = [32, 64, 128, 512]
ranges_list = [(0,128), (50,200), (100,256)]

for bins in bins_list:
    for r in ranges_list:
        print(f"\n=== Histogram with bins={bins}, range={r} ===")
        plot_histograms(img, equalized_img, bins, r)
```
