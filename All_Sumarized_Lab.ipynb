```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("pic.jpg", 0)

dft = np.fft.fft2(img)
dft_shift = np.fft.fftshift(dft)

rows, cols = img.shape
crow, ccol = rows//2, cols//2
mask = np.zeros((rows, cols), np.uint8)
mask[crow-30:crow+30, ccol-30:ccol+30] = 1

filtered = dft_shift * mask
f_ishift = np.fft.ifftshift(filtered)
img_back = np.abs(np.fft.ifft2(f_ishift))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.show()
plt.imshow(img_back, cmap='gray')
plt.title("Low-Pass Filtered")
plt.show()
```

```python
img = cv2.imread("pic.jpg", 0)
dft = np.fft.fft2(img)
dft_shift = np.fft.fftshift(dft)

rows, cols = img.shape
crow, ccol = rows//2, cols//2
mask = np.ones((rows, cols), np.uint8)
mask[crow-20:crow+20, ccol-20:ccol+20] = 0

filtered = dft_shift * mask
f_ishift = np.fft.ifftshift(filtered)
hp_img = np.abs(np.fft.ifft2(f_ishift))

plt.imshow(hp_img, cmap='gray')
plt.title("High-Pass Filtered")
plt.show()
```

```python
img = cv2.imread("pic.jpg", 0)

plt.hist(img.ravel(), bins=256, range=[0,256])
plt.title("Histogram")
plt.show()
```

```python
img = cv2.imread("pic.jpg", 0)
eq = cv2.equalizeHist(img)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(eq, cmap='gray'); plt.title("Equalized")
plt.show()
```

```python
img = cv2.imread("pic.jpg", 0)


noise = np.random.normal(0, 20, img.shape)
noisy = img + noise
noisy = np.clip(noisy, 0, 255).astype('uint8')

restored = cv2.GaussianBlur(noisy, (5,5), 0)

plt.subplot(1,2,1); plt.imshow(noisy, cmap='gray'); plt.title("Noisy")
plt.subplot(1,2,2); plt.imshow(restored, cmap='gray'); plt.title("Restored")
plt.show()
```

```python
img = cv2.imread("pic.jpg")
b, g, r = cv2.split(img)

plt.figure(figsize=(10,3))
plt.subplot(1,3,1); plt.imshow(r, cmap='gray'); plt.title("Red")
plt.subplot(1,3,2); plt.imshow(g, cmap='gray'); plt.title("Green")
plt.subplot(1,3,3); plt.imshow(b, cmap='gray'); plt.title("Blue")
plt.show()
```

```python
img = cv2.imread("pic.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.subplot(1,2,1); plt.imshow(gray, cmap='gray'); plt.title("Grayscale")
plt.subplot(1,2,2); plt.imshow(hsv[:,:,0], cmap='gray'); plt.title("Hue Channel")
plt.show()
```
