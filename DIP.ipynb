```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


image_path = "/215490-Trees-Of-Mystery-Redwood.webp"  # Change this to your image file path

image_bgr = cv2.imread(image_path)
if image_bgr is None:
    raise FileNotFoundError(f"Image not found at {image_path}. Please check the path.")

image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

bright_rgb = cv2.convertScaleAbs(image_rgb, alpha=1, beta=50)  # beta adds brightness
bright_gray = cv2.convertScaleAbs(gray, alpha=1, beta=50)

invert_rgb = 255 - image_rgb
invert_gray = 255 - gray

region_rgb = image_rgb.copy()
region_rgb[50:150, 50:150] = [255, 0, 0]  # set square region to red

region_gray = gray.copy()
region_gray[50:150, 50:150] = 255

fig, axes = plt.subplots(3, 3, figsize=(12, 10))

axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title("Original RGB")
axes[0, 0].axis("off")

axes[0, 1].imshow(bright_rgb)
axes[0, 1].set_title("Brightened RGB")
axes[0, 1].axis("off")

axes[0, 2].imshow(invert_rgb)
axes[0, 2].set_title("Inverted RGB")
axes[0, 2].axis("off")

axes[1, 0].imshow(gray, cmap="gray")
axes[1, 0].set_title("Original Grayscale")
axes[1, 0].axis("off")

axes[1, 1].imshow(bright_gray, cmap="gray")
axes[1, 1].set_title("Brightened Grayscale")
axes[1, 1].axis("off")

axes[1, 2].imshow(invert_gray, cmap="gray")
axes[1, 2].set_title("Inverted Grayscale")
axes[1, 2].axis("off")

axes[2, 0].imshow(region_rgb)
axes[2, 0].set_title("Region Manipulated RGB")
axes[2, 0].axis("off")

axes[2, 1].imshow(region_gray, cmap="gray")
axes[2, 1].set_title("Region Manipulated Gray")
axes[2, 1].axis("off")

axes[2, 2].axis("off")  # empty slot for cleaner layout

plt.tight_layout()
plt.show()
```
