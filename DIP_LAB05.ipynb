```python
import cv2

cap = cv2.VideoCapture("/content/yess.mp4")

cv2.namedWindow("All", cv2.WINDOW_NORMAL)  # Create window once

while cap.isOpened():
    r, frame = cap.read()
    if not r:
        break

    frame = cv2.resize(frame, (600, 500))
    cv2.imshow("All", frame)

    if cv2.waitKey(30) & 0xff == ord("p"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture("/content/yess.mp4")

while cap.isOpened():
    r, frame = cap.read()
    if r == True:
        frame = cv2.resize(frame, (600, 500))
        low_pass = cv2.GaussianBlur(frame, (7, 7), 0)
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        laplacian1 = cv2.Laplacian(gray, cv2.CV_8U)
        laplacian2 = cv2.Laplacian(gray, cv2.CV_64F)
        laplacian3 = cv2.convertScaleAbs(laplacian2)
        cv2_imshow(frame)
        cv2_imshow(low_pass)
        cv2_imshow(laplacian1)
        cv2_imshow(laplacian3)
        if cv2.waitKey(30) & 0xff == ord("p"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
